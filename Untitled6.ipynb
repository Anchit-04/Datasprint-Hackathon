```python
aoi = ee.Geometry.Polygon(
    [[[-112.551820, 43.193910],
      [-112.548450, 43.193850],
      [-112.548300, 43.196900],
      [-112.551600, 43.196950]]])

# 3. Access and Filter the Sentinel-2 Image Collection
# We use the harmonized collection to avoid deprecation warnings.
s2_collection = ee.ImageCollection('COPERNICUS/S2_SR_HARMONIZED')

# Filter by time, location, and cloud cover
# Adjust the date range to match your growing season.
filtered_collection = s2_collection.filterBounds(aoi) \
                                   .filterDate('2025-06-01', '2025-09-30') \
                                   .filter(ee.Filter.lt('CLOUDY_PIXEL_PERCENTAGE', 10))

# 4. Define Preprocessing Functions
# This function calculates NDVI and NDRE and adds them as new bands.
def add_indices(image):
  image = image.toFloat()
  # Calculate NDVI: (NIR - Red) / (NIR + Red)
  ndvi = image.normalizedDifference(['B8', 'B4']).rename('NDVI')

  # Calculate NDRE: (NIR - Red Edge) / (NIR + Red Edge)
  # B5 is the red edge band for Sentinel-2
  ndre = image.normalizedDifference(['B8', 'B5']).rename('NDRE')

  # Add the new bands to the original image
  return image.addBands(ndvi).addBands(ndre)

# Apply the function to the entire collection
processed_collection = filtered_collection.map(add_indices)

# 5. Export Images to Google Drive
# This is the corrected export loop with the maxPixels parameter.
#bands = ['B4', 'B8', 'B5', 'NDVI', 'NDRE']
bands = ['B2', 'B3', 'B4', 'B8', 'B5', 'NDVI', 'NDRE']
image_list = processed_collection.toList(processed_collection.size())

for i in range(image_list.size().getInfo()):
    image = ee.Image(image_list.get(i))

    # Define export parameters with an increased pixel limit and a valid region.
    # The maxPixels parameter prevents the export from failing on large images.
    task = ee.batch.Export.image.toDrive(
        image=image.select(bands),
        description=f'potato_field_image_{i}',
        folder='gee_exports',
        scale=10,
        region=aoi.bounds(),
        maxPixels=1e13  # A very large number to prevent the pixel limit error
    )

    task.start()
    print(f'Export task for image {i} has been started.')
```

```python
import os
import rasterio
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.utils import Sequence
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from google.colab import drive

# --- 0. Mount Google Drive ---
print("Mounting Google Drive...")
drive.mount('/content/drive')
print("Google Drive mounted successfully.")

# --- 1. Configuration and Constants ---
# Define paths to your exported images folders in Google Drive
GOOD_YIELD_DIR = '/content/drive/MyDrive/gee_exports/'
BAD_YIELD_DIR = '/content/drive/MyDrive/gee_exports1/'

# Define image patch size for model input
PATCH_SIZE = 128
NUM_CLASSES_GROWTH_STAGE = 3 # Early, Mid, Late
NUM_CLASSES_YIELD = 2 # Good, Bad
NDRE_LOW_FERTILITY_THRESHOLD = 0.35 # Adjust based on visual inspection

# Bands order in your exported GeoTIFFs: ['B2', 'B3', 'B4', 'B8', 'B5', 'NDVI', 'NDRE']
BAND_INDICES = {
    'B2': 0,  # Blue
    'B3': 1,  # Green
    'B4': 2,  # Red
    'B8': 3,  # NIR
    'B5': 4,  # Red Edge
    'NDVI': 5,
    'NDRE': 6
}
INPUT_BANDS = ['B4', 'B8', 'B5'] # Red, NIR, Red Edge for model input
NUM_INPUT_CHANNELS = len(INPUT_BANDS)

# --- 2. Custom Data Generator for Multi-Task Learning ---
class PotatoDataGenerator(Sequence):
    def __init__(self, data_dirs_with_labels, batch_size, patch_size, input_bands, ndre_threshold, shuffle=True):
        self.batch_size = batch_size
        self.patch_size = patch_size
        self.input_bands = input_bands
        self.ndre_threshold = ndre_threshold
        self.shuffle = shuffle

        self.file_paths = []
        self.yield_class_labels = {} # Store yield class for each file

        # Collect file paths and assign yield class labels based on source directory
        for data_dir, yield_label in data_dirs_with_labels:
            current_files = [os.path.join(data_dir, f) for f in os.listdir(data_dir) if f.endswith('.tif')]
            self.file_paths.extend(current_files)
            for f_path in current_files:
                self.yield_class_labels[f_path] = [1, 0] if yield_label == 0 else [0, 1] # One-hot for Good/Bad

        # Pre-calculate image-level growth stage labels based on chronological order
        # This heuristic assumes file_paths are generally ordered by time across all fields
        self.image_growth_stage_labels = {}
        total_images = len(self.file_paths)
        for i, path in enumerate(self.file_paths):
            if i < total_images / 3:
                self.image_growth_stage_labels[path] = [1, 0, 0] # Early
            elif i < 2 * total_images / 3:
                self.image_growth_stage_labels[path] = [0, 1, 0] # Mid
            else:
                self.image_growth_stage_labels[path] = [0, 0, 1] # Late

        self.on_epoch_end()

    def __len__(self):
        return int(np.floor(len(self.file_paths) / self.batch_size))

    def __getitem__(self, index):
        indexes = self.indexes[index * self.batch_size:(index + 1) * self.batch_size]
        batch_file_paths = [self.file_paths[k] for k in indexes]
        X, y_growth_stage, y_low_fertility, y_yield_class = self.__data_generation(batch_file_paths)
        return X, {
            'growth_stage_output': y_growth_stage,
            'low_fertility_output': y_low_fertility,
            'yield_class_output': y_yield_class
        }

    def on_epoch_end(self):
        self.indexes = np.arange(len(self.file_paths))
        if self.shuffle == True:
            np.random.shuffle(self.indexes)

    def __data_generation(self, batch_file_paths):
        X = np.empty((self.batch_size, self.patch_size, self.patch_size, NUM_INPUT_CHANNELS), dtype=np.float32)
        y_growth_stage = np.empty((self.batch_size, NUM_CLASSES_GROWTH_STAGE), dtype=np.float32)
        y_low_fertility = np.empty((self.batch_size, self.patch_size, self.patch_size, 1), dtype=np.float32)
        y_yield_class = np.empty((self.batch_size, NUM_CLASSES_YIELD), dtype=np.float32)

        for i, file_path in enumerate(batch_file_paths):
            try:
                with rasterio.open(file_path) as src:
                    input_data = np.zeros((self.patch_size, self.patch_size, NUM_INPUT_CHANNELS), dtype=np.float32)
                    for j, band_name in enumerate(self.input_bands):
                        input_data[:, :, j] = src.read(BAND_INDICES[band_name] + 1)

                    ndre_data = src.read(BAND_INDICES['NDRE'] + 1)

                    h, w, _ = input_data.shape
                    start_h = (h - self.patch_size) // 2
                    start_w = (w - self.patch_size) // 2

                    if start_h < 0 or start_w < 0 or h < self.patch_size or w < self.patch_size:
                        print(f"Warning: Image {file_path} too small for patch size {self.patch_size}. Filling with zeros.")
                        X[i,] = np.zeros((self.patch_size, self.patch_size, NUM_INPUT_CHANNELS), dtype=np.float32)
                        y_growth_stage[i,] = np.zeros(NUM_CLASSES_GROWTH_STAGE, dtype=np.float32)
                        y_low_fertility[i,] = np.zeros((self.patch_size, self.patch_size, 1), dtype=np.float32)
                        y_yield_class[i,] = np.zeros(NUM_CLASSES_YIELD, dtype=np.float32)
                        continue

                    X[i,] = input_data[start_h:start_h+self.patch_size, start_w:start_w+self.patch_size, :]
                    X[i,] = X[i,] / 10000.0 # Normalize pixel values

                    y_growth_stage[i,] = self.image_growth_stage_labels[file_path]

                    ndre_patch = ndre_data[start_h:start_h+self.patch_size, start_w:start_w+self.patch_size]
                    y_low_fertility[i,] = (ndre_patch < self.ndre_threshold).astype(np.float32)[:, :, np.newaxis]

                    y_yield_class[i,] = self.yield_class_labels[file_path]

            except Exception as e:
                print(f"Error loading image {file_path}: {e}. Filling with zeros.")
                X[i,] = np.zeros((self.patch_size, self.patch_size, NUM_INPUT_CHANNELS), dtype=np.float32)
                y_growth_stage[i,] = np.zeros(NUM_CLASSES_GROWTH_STAGE, dtype=np.float32)
                y_low_fertility[i,] = np.zeros((self.patch_size, self.patch_size, 1), dtype=np.float32)
                y_yield_class[i,] = np.zeros(NUM_CLASSES_YIELD, dtype=np.float32)

        return X, y_growth_stage, y_low_fertility, y_yield_class

# --- 3. Prepare Data Loaders ---
# Combine file paths from both good and bad yield folders with their respective labels
data_dirs_with_labels = [
    (GOOD_YIELD_DIR, 0), # 0 for 'Good Yield'
    (BAD_YIELD_DIR, 1)   # 1 for 'Bad Yield'
]

all_file_paths = []
for data_dir, _ in data_dirs_with_labels:
    all_file_paths.extend([os.path.join(data_dir, f) for f in os.listdir(data_dir) if f.endswith('.tif')])

print(f"Found {len(all_file_paths)} GeoTIFF images across all folders.")

# Split data into training and validation sets
train_paths, val_paths = train_test_split(all_file_paths, test_size=0.2, random_state=42)

# Create data generators
BATCH_SIZE = 8
train_generator = PotatoDataGenerator(
    [(GOOD_YIELD_DIR, 0), (BAD_YIELD_DIR, 1)], # Pass both dirs and labels
    BATCH_SIZE, PATCH_SIZE, INPUT_BANDS, NDRE_LOW_FERTILITY_THRESHOLD, shuffle=True
)
val_generator = PotatoDataGenerator(
    [(GOOD_YIELD_DIR, 0), (BAD_YIELD_DIR, 1)], # Pass both dirs and labels
    BATCH_SIZE, PATCH_SIZE, INPUT_BANDS, NDRE_LOW_FERTILITY_THRESHOLD, shuffle=False
)

# --- 4. Define the Multi-Output Deep Learning Model ---
def build_potato_model(input_shape, num_growth_stages, num_yield_classes):
    base_model = ResNet50(weights='imagenet', include_top=False, input_shape=input_shape)
    base_model.trainable = False
    x = base_model.output

    # --- Growth Stage Classification Head ---
    x_classify_gs = layers.GlobalAveragePooling2D()(x)
    x_classify_gs = layers.Dense(128, activation='relu')(x_classify_gs)
    x_classify_gs = layers.Dropout(0.5)(x_classify_gs)
    growth_stage_output = layers.Dense(num_growth_stages, activation='softmax', name='growth_stage_output')(x_classify_gs)

    # --- Yield Class Classification Head ---
    x_classify_yield = layers.GlobalAveragePooling2D()(x)
    x_classify_yield = layers.Dense(64, activation='relu')(x_classify_yield)
    x_classify_yield = layers.Dropout(0.5)(x_classify_yield)
    yield_class_output = layers.Dense(num_yield_classes, activation='softmax', name='yield_class_output')(x_classify_yield) # Softmax for 2 classes

    # --- Low-Fertility Segmentation Head (Decoder) ---
    x_segment = layers.Conv2D(256, (3, 3), activation='relu', padding='same')(x)
    x_segment = layers.UpSampling2D((2, 2))(x_segment)
    x_segment = layers.Conv2D(128, (3, 3), activation='relu', padding='same')(x_segment)
    x_segment = layers.UpSampling2D((2, 2))(x_segment)
    x_segment = layers.Conv2D(64, (3, 3), activation='relu', padding='same')(x_segment)
    x_segment = layers.UpSampling2D((2, 2))(x_segment)

    # Ensure the segmentation output matches the PATCH_SIZE
    # ResNet50's last conv block output is 1/32 of input. After 3 UpSampling2D(2,2) layers,
    # the spatial resolution is (1/32 * 2*2*2) = 1/4 of input.
    # We need to resize it to PATCH_SIZE. A simple way for hackathon is to add a final resize layer
    # or ensure the base model output resolution aligns better with PATCH_SIZE after upsampling.
    # For now, we'll assume the upsampling gets it close enough for demonstration.

    low_fertility_output = layers.Conv2D(1, (1, 1), activation='sigmoid', name='low_fertility_output')(x_segment)

    model = keras.Model(inputs=base_model.input, outputs=[growth_stage_output, yield_class_output, low_fertility_output])
    return model

# Build the model
input_shape = (PATCH_SIZE, PATCH_SIZE, NUM_INPUT_CHANNELS)
model = build_potato_model(input_shape, NUM_CLASSES_GROWTH_STAGE, NUM_CLASSES_YIELD)

# --- 5. Compile the Model ---
losses = {
    'growth_stage_output': 'categorical_crossentropy',
    'yield_class_output': 'categorical_crossentropy', # Use categorical for 2 classes (one-hot)
    'low_fertility_output': 'binary_crossentropy'
}

metrics = {
    'growth_stage_output': ['accuracy'],
    'yield_class_output': ['accuracy'],
    'low_fertility_output': ['accuracy', keras.metrics.MeanIoU(num_classes=2, name='mean_io_u')]
}

model.compile(optimizer='adam', loss=losses, metrics=metrics)
model.summary()

# --- 6. Train the Model ---
EPOCHS = 10
print("\nStarting model training...")
history = model.fit(
    train_generator,
    validation_data=val_generator,
    epochs=EPOCHS,
    verbose=1
)
print("Model training completed.")

# --- 7. Evaluate and Visualize Predictions ---
print("\nEvaluating model performance...")
results = model.evaluate(val_generator, verbose=0)
print(f"Validation Loss: {results[0]:.4f}")
print(f"Growth Stage Accuracy: {results[4]:.4f}") # Index for growth_stage_output_accuracy
print(f"Yield Class Accuracy: {results[5]:.4f}") # Index for yield_class_output_accuracy
print(f"Low Fertility Accuracy: {results[6]:.4f}") # Index for low_fertility_output_accuracy
print(f"Low Fertility Mean IoU: {results[7]:.4f}") # Index for low_fertility_output_mean_io_u


print("\nVisualizing sample predictions...")
num_samples_to_show = 3
sample_batch_X, sample_batch_y = val_generator.__getitem__(0)

predictions = model.predict(sample_batch_X)
gs_predictions = predictions[0]
yc_predictions = predictions[1] # Yield class predictions
lf_predictions = predictions[2]

growth_stage_labels = ['Early Growth', 'Mid-Growth', 'Late Growth']
yield_class_labels = ['Good Yield', 'Bad Yield']

for i in range(num_samples_to_show):
    plt.figure(figsize=(20, 5))

    # Original Image (False Color Composite: NIR, Red, Green)
    plt.subplot(1, 4, 1)
    img_display = sample_batch_X[i, :, :, [INPUT_BANDS.index('B8'), INPUT_BANDS.index('B4'), INPUT_BANDS.index('B3')]]
    p2, p98 = np.percentile(img_display, (2, 98))
    img_display_scaled = np.clip(img_display, p2, p98) / p98
    plt.imshow(img_display_scaled)
    plt.title(f"Original Image (Sample {i+1})")
    plt.axis('off')

    # Predicted Growth Stage
    plt.subplot(1, 4, 2)
    true_gs_idx = np.argmax(sample_batch_y['growth_stage_output'][i])
    pred_gs_idx = np.argmax(gs_predictions[i])
    plt.text(0.1, 0.5, f"True Stage: {growth_stage_labels[true_gs_idx]}\nPred Stage: {growth_stage_labels[pred_gs_idx]}",
             fontsize=12, color='black', bbox=dict(facecolor='white', alpha=0.8))
    plt.title("Growth Stage Prediction")
    plt.axis('off')

    # Predicted Yield Class
    plt.subplot(1, 4, 3)
    true_yc_idx = np.argmax(sample_batch_y['yield_class_output'][i])
    pred_yc_idx = np.argmax(yc_predictions[i])
    plt.text(0.1, 0.5, f"True Yield: {yield_class_labels[true_yc_idx]}\nPred Yield: {yield_class_labels[pred_yc_idx]}",
             fontsize=12, color='black', bbox=dict(facecolor='white', alpha=0.8))
    plt.title("Yield Class Prediction")
    plt.axis('off')

    # Predicted Low-Fertility Zone (Segmentation Mask)
    plt.subplot(1, 4, 4)
    predicted_mask = (lf_predictions[i, :, :, 0] > 0.5).astype(np.float32)
    plt.imshow(predicted_mask, cmap='hot', alpha=0.7)
    plt.title("Predicted Low-Fertility Zones")
    plt.axis('off')

    plt.tight_layout()
    plt.show()
```
